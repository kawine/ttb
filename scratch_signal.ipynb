```python
from ttb import utils
import numpy as np
import random
from matplotlib import pyplot as plt
```

```python
def create_domain_matrix(n, v):
    """
    Args:
        n: number of examples
        v: number of domain values
    """
    X = np.zeros((n, v))
    indices = [ (i, np.random.randint(0, v)) for i in range(n) ]
    rows, cols = zip(*indices)
    X[rows, cols] = 1
    return X

def visualize_signal(signal, domain_type_index, start, end, T):
    for i in range(start, end):
        x = list(range(T))
        y = [ signal[t][domain_type_index][i] for t in x ]
        plt.plot(x, y)
    
    plt.ylim([-1, 10])
    plt.rcParams["figure.figsize"] = (20,4)
    plt.show()
    
def visualize_prob(p, i):
    x = list(range(T))
    y = np.array(p)[:,i]
    plt.plot(x, y)
    plt.show()
```

```python
num_examples = 100
num_domain_values = [ 5, 4, 2 ]
T = 200

domain_matrices = [ create_domain_matrix(num_examples, v) for v in num_domain_values ]
probs, signals = utils.create_probabilities(
    domain_matrices,
    num_examples,
    T,
    gamma=0.5,  # good default choice
    num_peaks=5, # good default choice
    log_step=10,
)
```

```python
visualize_signal(signals, 0, 0, 2, T)
```

```python
visualize_signal(signals, 1, 0, 3, T)
```

```python
visualize_signal(signals, 2, 0, 2, T)
```

```python
visualize_prob(probs, 10)
```
